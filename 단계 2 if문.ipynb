```python
#1 두 수 비교하기
A, B = map(int, input().split())
if (A > B) :
    print(">")
elif (A < B) :
    print("<")
else :
    print("==")
    
```

```python
#4 사분면

a = int(input())
b = int(input())

if a>0 and b>0 :
    print(1)
elif a<0 and b>0 :
    print(2)
elif a<0 and b<0 :
    print(3)
else :
    print(4)
```
